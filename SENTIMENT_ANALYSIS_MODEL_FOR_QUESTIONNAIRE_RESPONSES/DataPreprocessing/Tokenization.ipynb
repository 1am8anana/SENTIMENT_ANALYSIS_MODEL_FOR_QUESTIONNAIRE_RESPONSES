# Tokenization

##### Import library

import pandas library to handle dataframe, Tokenizer for make the sentence to tokens.

```python
import pandas as pd
from CustomTokenizer import Tokenizer
```

##### Load the dataset

```python
datasets = pd.read_csv('../Dataset/datasets.csv')
# Get the sentences
sentences = datasets['sentence']
```

##### Tokenize the sentence

```python
# Instantiate the Tokenizer.
tokenizer = Tokenizer()
# Fit the sentences and transforms it to the tokesn.
str_tokens = tokenizer.fit_transform(sentences)
```
