# Feature Extractions

##### Import Library

we need to import some library like pandas for handle dataframe, Customtokenizer (that we have created) to make the tokens, and Vectorizer that we want to have TF-IDF matrix.

```python
import pandas as pd
from CustomTokenizer import Tokenizer
from Vectorizer import Vectorizer
```

Load the (sentences)feature

```python
dataset = pd.read_csv('../Dataset/datasets.csv')
sentences = dataset['sentence']
```

```python
# Instantiate the tokenizer
tokenizer = Tokenizer()
# fit and transform the sentences to the tokenizer.
str_tokens = tokenizer.fit_transform(sentences)
# Now we have tokens.
```

```python
str_tokens
```

TF-IDF matrix

```python
# Instantiate (Tfidf)vectorizer object.
vectorizer = Vectorizer()
# fit the sentences to Create TF-IDF.
vectorizer.fit(str_tokens)
```

##### Explore the tokens, Vocabulary and The TF-IDF matrix

```python
vectorizer.vocabulary_
```

Let's see the matrix. In this case it hard to see all of the values

```python
transform_data = vectorizer.transform(sentences)
pd.DataFrame(transform_data.toarray())
```

From this results aboved we have:
* 342 raws that refers to sentences that just fitted in the tokenizer and vectorizer.
* 1128 columns this refers to words/tokens that we have (in the vocabulary/TF-IDF matrix).
